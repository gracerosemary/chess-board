```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid = np.ones((8, 8, 3)) # fill it with ones - np.zeros fills with 0s (makes black)
# help(np.ones)
grid
```

```python
# help(plt.imshow)
# grid[0][0] = (1,1,1) # changes 1 block to white
# grid[1][1] = (1,1,1) # changes next diagonal block to white
white = (1,1,1)
for i in range(8):
    grid[i][i] = white
plt.imshow(grid)
```

```python
class ChessBoard:
    def __init__(self):
        self.grid = np.zeros((8,8,3))
        white = [1,1,1]
        for i in range(8):
            for j in range(8):
                if (i+j)%2:
                    self.grid[i][j] = white

    red = []
    def add_red(self, row, column):
        red_coordinates = []
        red_coordinates.append(row)
        red_coordinates.append(column)
        ChessBoard.red.append(red_coordinates)
        self.grid[row, column] = [1, 0, 0]

    blue = []
    def add_blue(self, row, column):
        blue_coordinates = []
        blue_coordinates.append(row)
        blue_coordinates.append(column)
        ChessBoard.blue.append(blue_coordinates)
        self.grid[row, column] = [0, 0, 1]

    def render(self):
        plt.imshow(self.grid)

    def check_slope(self):
        rise = ChessBoard.blue[0][1] - ChessBoard.red[0][1]
        run = ChessBoard.blue[0][0] - ChessBoard.red[0][0]
        return rise / run

    def is_under_attack(self):
        slope = ChessBoard.check_slope(self)
        print(slope)
        if slope == 1 or slope == -1:
            return True
        else: 
            return False

if __name__ == "__main__":
    cb = ChessBoard()
    cb.add_red(5,6)
    cb.add_blue(6,7)
    cb.render()
    print(cb.check_slope())
    print(cb.is_under_attack())
```
